````python
# ============================================================
# Step 3 ‚Äî Display predicted results and plot
# ============================================================

import pandas as pd
import matplotlib.pyplot as plt

try:
    data = pd.read_csv("predicted_results.csv")
    print("‚úÖ Predicted Results:\n")
    display(data.head())

    plt.figure(figsize=(6,4))
    plt.scatter(data["Predicted_Eg(eV)"], data["Predicted_CO2_Efficiency(%)"], s=80, color='green')
    plt.xlabel("Predicted Bandgap (eV)")
    plt.ylabel("Predicted CO‚ÇÇ Conversion Efficiency (%)")
    plt.title("Predicted Bandgap vs CO‚ÇÇ Conversion Efficiency (Anderson-type POMs)")
    plt.grid(True)
    plt.show()

except FileNotFoundError:
    print("‚ö†Ô∏è No predicted_results.csv file found. Check if HybridCatPredictor.py outputs it correctly.")
````
